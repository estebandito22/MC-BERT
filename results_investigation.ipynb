```python
import csv

answer_file = "mscoco_val2014_answers.csv"
result_file = "results.txt"
labels_file = "answer_labels.txt"

question_info = {}

uniq_question_types = set()
uniq_answer_types = set()

with open(answer_file, "r") as f:

    reader = csv.reader(f)

    for row in reader:
        qid = row[0]
        best_answer = row[1]
        question_type = row[2]
        answer_type = row[3]
        top10_answers = row[4:]

        uniq_question_types.add(question_type)
        uniq_answer_types.add(answer_type)

        question = {'answer':best_answer, 'qtype': question_type, 'atype':answer_type}

        question_info[qid] = question
    
total_questions = len(question_info)
    
results_info = {}

answers_correct = {}
answers_total = {}
questions_correct = {}
questions_total = {}

for ans in uniq_answer_types:
    answers_correct[ans] = 0
    answers_total[ans] = 0

for qes in uniq_question_types:
    questions_correct[qes] = 0
    questions_total[qes] = 0
    
total_correct = 0

answers = []

with open(answer_file, "r") as f:
    for res in f:
        answers.append(res.strip())

#Use this when we're marking classes > 3000 as "unknown"
answers[3000] = "UNKNOWN"

with open(result_file, "r") as f:

    reader = csv.reader(f)
    
    for row in reader:
        qid = row[0]
        correct = (row[1] == row[2])
        
        total_correct += correct
        
        question = question_info[qid]
        answers_correct[question['atype']] += correct
        answers_total[question['atype']] += 1
        questions_correct[question['qtype']] += correct
        questions_total[question['qtype']] += 1
        question['prediction'] = answers[int(row[1])]
        

print("Total Accuracy: {:2.2f}%".format(total_correct/ total_questions * 100))
print()      
```

```python

for ans in uniq_answer_types:
    print(ans, "({:2.0f}%): ".format(answers_total[ans] / total_questions * 100), "{:2.2f}%".format(answers_correct[ans] / answers_total[ans] * 100), sep='' )
```

```python
qes_correct = []
for qes in uniq_question_types:
    qes_correct.append((qes,questions_correct[qes] / questions_total[qes] * 100 ))
    
    
qes_correct = sorted(qes_correct, key=lambda x: x[1], reverse=True)
for qes in qes_correct:
    print(qes[0], "({}): ".format(questions_total[qes[0]]), "{:2.2f}%".format(qes[1]), sep='')
    
```
